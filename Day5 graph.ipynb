```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
```

```python
plt.plot(x,y)
plt.title("students hrs vs marks ")
plt.xlabel("Marks")
plt.ylabel("Hrs")
plt.show()
```

```python
subject = ["M1","M2","Chemistry","Physics"]
marks = ["15","19","16","20"]
plt.bar(subject,marks)
plt.title("students Subjects vs marks ")
plt.xlabel("Marks")
plt.ylabel("Subjects")
plt.show()

```
